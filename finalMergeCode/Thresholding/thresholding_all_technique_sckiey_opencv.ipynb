# Using Sckiey

```python
from skimage import io

from skimage.filters import try_all_threshold

from skimage.filters import threshold_otsu

from skimage.filters.rank import entropy

import matplotlib.pyplot as plt
```

```python
img = io.imread("demo/demo0.jpg", as_gray=True)
```

## Different Types Of Thresholding

```python
fig, ax = try_all_threshold(img, figsize=(10, 8), verbose=False)
plt.show()
```

## Otsu Thresholding

```python
thresh = threshold_otsu(img)
binary= img <=thresh   
plt.imshow(binary, cmap=plt.cm.gray)
```

# Using OpenCv

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inLine
```

```python
img =cv2.imread("demo/demo0.jpg",0)
```

```python
plt.imshow(img)
```

## Coverting BGR Image To RGB

```python
img_0=cv2.cvtColor(img,cv2.COLOR_BGR2RGB) 
```

```python
plt.imshow(img_0)
```

## Thresholding Binary

```python
ret,thresh1=cv2.threshold(img,127,229,cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh1)
```

## Inverse Thresholding Binary

```python
ret_0,thresh1_0=cv2.threshold(img,127,229,cv2.THRESH_BINARY_INV)
```

```python
plt.imshow(thresh1_0)
```
