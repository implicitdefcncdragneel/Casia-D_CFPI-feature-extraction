```python
def show_pic(img):
    fig=plt.figure(figsize=(15,15))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap="gray")
```

```python
show_pic(img)
```

```python
ret_0_0_0_1,thresh1_0_0_0_1=cv2.threshold(img,127,229,cv2.THRESH_BINARY)
show_pic(thresh1_0_0_0_1)
```

```python
#################################################
#Adaptive Thresholding
#################################################
```

```python
th2=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
show_pic(th2)
```

```python
th3=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,8)
show_pic(th3)
```

```python
blended_img=cv2.addWeighted(src1=th2,alpha=0.6,src2=th3,beta=0.4,gamma=0)
show_pic(blended_img)
```
