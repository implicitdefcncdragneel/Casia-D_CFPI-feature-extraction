```python
import cv2

import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def load_img():
    img = cv2.imread('datasetPerson24/demo3.jpg').astype(np.float32) / 255
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
#load_img()
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
i = load_img()
display_img(i)
```

```python
gamma=1/4
```

```python
result=np.power(i,gamma)
display_img(result)
```

```python
img_0 = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img_0,text='CNC',org=(1,50),fontFace=font,fontScale=1,color=(0,0,255),thickness=1)
display_img(img_0)
```

```python
kernel = np.ones(shape=(5,5),dtype=np.float32)/25
```

```python
kernel
```

```python
dst = cv2.filter2D(img_0,-1,kernel)
display_img(dst)
```

```python
img_0_1 = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img_0_1,text='CNC',org=(1,50),fontFace=font,fontScale=1,color=(0,0,255),thickness=1)
#display_img(img_0_1)
print("reset")
```

```python
blurred_img = cv2.blur(img_0_1,ksize=(5,5))
display_img(blurred_img)
```
