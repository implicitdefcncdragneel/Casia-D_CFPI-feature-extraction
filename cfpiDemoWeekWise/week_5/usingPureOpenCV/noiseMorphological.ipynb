```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    img =np.array(cv2.imread("datasetPerson24/demo6.jpg",0))
    return img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = load_img()
```

```python
display_img(img)
```

```python
kernel = np.ones((3,3),np.uint8)
```

```python
kernel
```

```python
erosion1 = cv2.erode(img,kernel,iterations = 1)
```

```python
display_img(erosion1)
```
