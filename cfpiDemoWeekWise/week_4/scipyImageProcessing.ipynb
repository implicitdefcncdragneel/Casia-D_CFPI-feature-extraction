# Geomatric Transformation

```python
flipped_img_LR = np.fliplr(img_0)
flipped_img_UD = np.flipud(img_0)
```

```python
plt.subplot(2,1,1)
plt.imshow(img_0, cmap="Greys")
plt.subplot(2,2,3)
plt.imshow(flipped_img_LR, cmap="Blues")
plt.subplot(2,2,4)
plt.imshow(flipped_img_UD, cmap="hsv")
```

```python
flipped_img_LR = np.fliplr(img_1)
flipped_img_UD = np.flipud(img_1)
```

```python
plt.subplot(2,1,1)
plt.imshow(img_1, cmap="Greys")
plt.subplot(2,2,3)
plt.imshow(flipped_img_LR, cmap="Blues")
plt.subplot(2,2,4)
plt.imshow(flipped_img_UD, cmap="hsv")
```

```python
rotated_img_0 = ndimage.rotate(img_0, 45)
plt.imshow(rotated_img_0)
```

```python
rotated_img_noreshape_0 = ndimage.rotate(img_0, 45, reshape=False)
plt.imshow(rotated_img_noreshape_0)
```

```python
rotated_img_1 = ndimage.rotate(img_1, 45)
plt.imshow(rotated_img_1)
```

```python
rotated_img_noreshape_1 = ndimage.rotate(img_1, 45, reshape=False)
plt.imshow(rotated_img_noreshape_1)
```

```python
uniform_filtered_img = ndimage.uniform_filter(img_0, size=9)
plt.imshow(uniform_filtered_img)
```

```python
uniform_filtered_img_1 = ndimage.uniform_filter(img_1, size=9)
plt.imshow(uniform_filtered_img_1)
```

# Applying Gaussian Filter for Smooth Noise And Edge Detection

```python
blurred_img = ndimage.gaussian_filter(img_0, sigma=3)
plt.imshow(blurred_img)
```

```python
blurred_img_1 = ndimage.gaussian_filter(img_1, sigma=3)
plt.imshow(blurred_img_1)
```

```python
median_img = ndimage.median_filter(img_0, 3)
plt.imshow(median_img)
```

```python
median_img_1 = ndimage.median_filter(img_1, 7)
plt.imshow(median_img_1)
```

# Edge Detection

```python
sobel_img = ndimage.sobel(img_0, axis=0)
plt.imshow(sobel_img)
```

```python
sobel_img_1 = ndimage.sobel(img_1, axis=0) 
plt.imshow(sobel_img_1)
```
